# 2. Как с помощью Markdown сформировать иерархический список?
Использовать отступы (пробелы или табуляцию) перед маркерами. Пример:
  
markdown
  - Уровень 1
    - Уровень 2,
      - Уровень 3
        1. Подпункт
Результат – многоуровневый (вложенный) список.
# 3. Как с помощью Markdown оформить цитату?
Поставить символ > в начале строки. Для вложенной цитаты – >>.
       
markdown
> Основная цитата,
>> Вложенная цитата

# 7. Типы данных (подробно об одном – числовые)
В Python/pandas: int, float, complex, bool, category, datetime64.
        
Числовые типы (int и float):
        
int – целые числа с фиксированной разрядностью (int8…int64). Используются для идентификаторов, счётчиков, возраста. Экономят память при малом диапазоне.
        
float – числа с плавающей точкой (float32, float64). float64 (двойная точность) занимает 8 байт, точность ~15–17 знаков. Необходимы для непрерывных величин (рост, вес, температура). Из-за неточного представления десятичных дробей сравнение выполняется через np.isclose().
        
complex – комплексные числа (редко в стандартном анализе).
